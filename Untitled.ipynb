```python
X = cancer_df.drop(['target'],axis = 1)
X.head(5)
```

```python
y = cancer_df['target']
y.head(6)
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.head(21)
```

```python
y_test[421]
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#XGBoost Classifier
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier()
xgb_classifier.fit(X_train, y_train)
y_pred_xgb = xgb_classifier.predict(X_test)
accuracy_score(y_test, y_pred_xgb)
```

```python
# Train with Standard scaled Data #kid and shape example
xgb_classifier2 = XGBClassifier()
xgb_classifier2.fit(X_train_sc, y_train)
y_pred_xgb_sc = xgb_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_xgb_sc)
```

```python
# XGBoost classifier most required parameters
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ] 
}
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_xgb_sc)

# Print the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
cm = confusion_matrix(y_test, y_pred_xgb_sc)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
print(classification_report(y_test, y_pred_xgb_sc))
```
